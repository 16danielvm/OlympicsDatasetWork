# Que dicen los datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('3 - Real_Olympics_Dataset.csv', index_col=[0])
df.head()
```

### RANKING DEL VERANO

```python
df_Winner_Summer = df[0:146].sort_values(['Gold_Summer','Silver_Summer','Bronze_Summer'], ascending=False)
df_Winner_Summer[['Pais', 'Codigo', 'Gold_Summer', 'Silver_Summer','Bronze_Summer']]
```

### RANKING DE INVIERNO

```python
df_Winner_Winter = df[0:146].sort_values(['Gold_Winter','Silver_Winter','Bronze_Winter'], ascending=False)
df_Winner_Winter[['Pais', 'Codigo', 'Gold_Winter', 'Silver_Winter','Bronze_Winter']]
```

### RANKING DE PAISES QUE MAS MEDALLAS DE VERANO A GANADO.

```python
df_Winner_Medals_Summer = df[0:146].sort_values(['Total_Summer'], ascending=False)
df_Winner_Medals_Summer[['Pais','Codigo','Total_Summer']]
```

### RANKING DE PAISES QUE MAS MEDALLAS DE INVIERNO A GANADO.

```python
df_Winner_Medals_Winter = df[0:146].sort_values(['Total_Winter'], ascending=False)
df_Winner_Medals_Winter[['Pais','Codigo','Total_Winter']]
```

### RANKING TOTAL

```python
df_Winner_Total = df[0:146].sort_values(['Gold_Total', 'Silver_Total', 'Bronze_Total'], ascending=False)
df_Winner_Total[['Pais', 'Codigo', 'Gold_Total', 'Silver_Total', 'Bronze_Total']]
```

### RANKING TOTAL DE MEDALLAS

```python
df_Winner_Medals_Total = df[0:146].sort_values(['Combined total'], ascending=False)
df_Winner_Medals_Total[['Pais','Codigo','Combined total']]
```

### RANKING DE PAISES QUE MAS HAN PARTICIPADO EN LOS JUEGOS DE VERANO

```python
df_Winner_GAMES_SUMMER = df[0:146].sort_values(['N_Summer'], ascending=False)
df_Winner_GAMES_SUMMER[['Pais','Codigo','N_Summer']]
```

### RANKING DE PAISES QUE MAS HAN PARTICIPADO EN LOS JUEGOS DE INVIERNO

```python
df_Winner_GAMES_WINTER = df[0:146].sort_values(['N_Winter'], ascending=False)
df_Winner_GAMES_WINTER[['Pais','Codigo','N_Winter']]
```

### RANKING DE PAISES QUE MAS HAN PARTICIPADO EN LOS JUEGOS OLIMPICOS EN TOTAL

```python
df_Winner_GAMES_TOTAL = df[0:146].sort_values(['N_Games'], ascending=False)
df_Winner_GAMES_TOTAL[['Pais','Codigo','N_Games']]
```

# IMPLEMENTACION VISUALIZACIONES

```python
df_Winner_Summer.head()
```

```python
fig, ax = plt.subplots()

ax.bar(df_Winner_Summer['Pais'][:5], df_Winner_Summer['Gold_Summer'][:5], label='Medalla de oro', )
ax.bar(df_Winner_Summer['Pais'][:5], df_Winner_Summer['Silver_Summer'][:5], label='Medalla de plata', bottom=df_Winner_Summer['Gold_Summer'][:5])
ax.bar(df_Winner_Summer['Pais'][:5], df_Winner_Summer['Bronze_Summer'][:5], label='Medalla de bronce', bottom=df_Winner_Summer['Gold_Summer'][:5] + df_Winner_Summer['Silver_Summer'][:5])

ax.set_title('Top 5 con mayor cantidad de medallas en verano')
ax.set_xlabel('Top 4 paises')
ax.set_ylabel('Cantidad de medallas')
plt.legend()

plt.show()
```

```python
import sys
fig, ax = plt.subplots(3,1, sharey=True, sharex=True)
ax[0].bar(df_Winner_Summer['Pais'][:5], df_Winner_Summer['Gold_Summer'][:5], color='y' )
ax[1].bar(df_Winner_Summer['Pais'][:5], df_Winner_Summer['Silver_Summer'][:5], color='b')
ax[2].bar(df_Winner_Summer['Pais'][:5], df_Winner_Summer['Bronze_Summer'][:5], color='r')

ax[0].set_title('Top 5 con mayor cantidad de medallas en verano')
ax[2].set_xticklabels(df_Winner_Summer['Pais'][:5], rotation=90) 
ax[0].set_ylabel('Cantidad\n de medallas\n de oro')
ax[1].set_ylabel('Cantidad\n de medallas\n de plata')
ax[2].set_ylabel('Cantidad\n de medallas\n de bronce')

plt.show()
```

```python
fig, ax  = plt.subplots(1,3, sharey=True, sharex=True)

ax[0].hist(df_Winner_Summer['N_Games'], bins=[x*5 for x in range(11)], histtype='step')
ax[0].set_ylabel("Cantidad de paises")
ax[0].set_xlabel("Cantidad de \n juegos totales")

ax[1].hist(df_Winner_Summer['N_Summer'], bins=[x*5 for x in range(7)],histtype='step')
ax[1].set_xlabel("Cantidad de \n juegos de verano")

ax[2].hist(df_Winner_Summer['N_Winter'], bins=[x*5 for x in range(6)],histtype='step')
ax[2].set_xlabel("Cantidad de juegos \n de invierno")


plt.show()
```

```python
fig, ax = plt.subplots()

ax.boxplot([df_Winner_Summer['N_Games'],df_Winner_Summer['N_Summer'], df_Winner_Summer['N_Winter']])
ax.set_xticklabels(['Juegos totales', 'Juegos de verano','Juegos de invierno'])
ax.set_ylabel('Numero de juegos')
ax.set_xlabel('Juegos Olimpicos')


plt.show()
```
